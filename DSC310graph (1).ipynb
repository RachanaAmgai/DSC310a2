```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
categories = ["A", "B", "C", "D"]
values1 = [10, 15, 20, 25]
values2 = [5, 10, 15, 20]

# 1. Vertical Bar Chart
plt.figure(figsize=(6, 4))
plt.bar(categories, values1, color='blue')
plt.title("Vertical Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()

# 2. Horizontal Bar Chart
plt.figure(figsize=(6, 4))
plt.barh(categories, values1, color='green')
plt.title("Horizontal Bar Chart")
plt.xlabel("Value")
plt.ylabel("Category")
plt.show()

# 3. Grouped Bar Chart
x = np.arange(len(categories))  # X locations for the groups
width = 0.35  # Width of the bars

plt.figure(figsize=(6, 4))
plt.bar(x - width/2, values1, width, label='Value1', color='blue')
plt.bar(x + width/2, values2, width, label='Value2', color='red')
plt.title("Grouped Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.xticks(x, categories)
plt.legend()
plt.show()

# 4. Stacked Bar Chart
plt.figure(figsize=(6, 4))
plt.bar(categories, values1, label='Value1', color='blue')
plt.bar(categories, values2, bottom=values1, label='Value2', color='red')
plt.title("Stacked Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.legend()
plt.show()

# 5. Heat Map
heatmap_data = np.array([[10, 20, 30, 40],
                         [15, 25, 35, 45],
                         [20, 30, 40, 50]])
rows = ["X", "Y", "Z"]
cols = ["A", "B", "C", "D"]

plt.figure(figsize=(6, 4))
plt.imshow(heatmap_data, cmap='hot', interpolation='nearest', aspect='auto')
plt.colorbar(label='Value')
plt.title("Heat Map")
plt.xticks(ticks=np.arange(len(cols)), labels=cols)
plt.yticks(ticks=np.arange(len(rows)), labels=rows)
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```
